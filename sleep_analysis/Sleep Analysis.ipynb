```python
# Import dependencies 
import pandas as pd
```

```python
# Read in the datset and select only the columns that are needed relating to sleep data
sleep_df = pd.read_csv('sleep_analysis.csv', usecols =['date','Sleep Score','Sleep Score','Total Sleep Score','Restfulness Score','Total Sleep Duration', 'Total Bedtime ', 'Awake Time','REM Sleep Duration', 'Light Sleep Duration', 'Deep Sleep Duration','Bedtime Start','Move Every Hour Score','Sleep Balance Score'])
```

```python
# Convert duration times from seconds to hours
sleep_df['Total Sleep Duration'] = sleep_df['Total Sleep Duration'] / 60**2
sleep_df['Awake Time'] = sleep_df['Awake Time'] / 60**2
sleep_df['Total Bedtime '] = sleep_df['Total Bedtime '] / 60**2
sleep_df['REM Sleep Duration'] = sleep_df['REM Sleep Duration'] / 60**2
sleep_df['Light Sleep Duration'] = sleep_df['Light Sleep Duration'] / 60**2
sleep_df['Deep Sleep Duration'] = sleep_df['Deep Sleep Duration'] / 60**2
```
